```python
import csv
```

```python
items = csv.reader(open('books.csv','r',encoding='utf-8'))
```

```python
for item in items:
    def entry():
        if item[0] == 'book':
            abbr = '<abbr type="siglum">' + item[1] + ' ' + item[3] + '</abbr>'
            author = '<author>'+item[1] + ', ' + item[2]+'</author>'
            title = '<title>'+item[4]+'</title>'
            volume = item[5]
            pubplace = '<pubPlace>'+item[6]+'</pubPlace>'
            date = '<date>'+item[3]+'</date>'

            wrapper_start = '<bibl xml:id="' + item[1]+item[3] + '">'
            wrapper_end = '</bibl>'

            if item[4] is '':
                citation = wrapper_start + abbr + ' ' + author + ' ' + title + ' ' + pubplace + ' ' + date + wrapper_end
                return citation
            else:
                citation = wrapper_start + abbr + ' ' + author + ' ' + title + ' ' + volume + ' ' + pubplace + ' ' + date + wrapper_end
                return citation
        else:
            
            abbr = '<abbr type="siglum">' + item[1] + ' ' + item[3] + '</abbr>'
            author = '<author>'+item[1] + ', ' + item[2] + '</author>'
            article_title = '<title level="a">'+item[4]+'</title>'
            journal_title = '<title level="j">'+item[5]+'</title>'
            volume = '<biblScope unit="volume">'+item[6]+'</biblScope>'
            pages = '<biblScope unit="page">'+item[7]+'</biblScope>'
            date = '<date>'+item[3]+'</date>'

            wrapper_start = '<bibl xml:id="' + item[1]+item[3] + '">'
            wrapper_end = '</bibl>'

            citation = wrapper_start + ' ' + abbr + ' ' + author + ' ' + article_title + ' ' + journal_title + ' ' + volume + '(' + date + '): ' + pages + wrapper_end
            
            return citation
    
    citation = entry()
    print(citation)
```
